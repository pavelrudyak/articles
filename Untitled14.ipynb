```python
X = []
y = []
for word in selected_words:
    word_index = model.wv.key_to_index[word.lower()]
    word_vector = model.wv.vectors[word_index]
    X.append(word_vector)
    y.append(word_classes[word])

X = np.array(X)
y = np.array(y)

# Создание и обучение KNN
knn = KNeighborsClassifier(n_neighbors=23)  # Устанавливаем 1 соседа
knn.fit(X, y)
```

```python
train = []
for i in data:
  train.append(data[i][-1])
```

```python
import numpy as np
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import Pipeline
```

```python
count_vect = CountVectorizer()
X_train_counts = count_vect.fit_transform(train)
```

```python
# Преобразуем матрицу количества токенов в нормализованное представление tf-idf (tf-idf transformer)
tfidf_transformer = TfidfTransformer()
X_train_tfidf = tfidf_transformer.fit_transform(X_train_counts)
```

```python
target = []
for i in data:
  target.append(data[i][0])
```

```python
np.shape(target)
```

```python
# Создаем классификатор K ближайших соседей с 7 соседями
knn = KNeighborsClassifier(n_neighbors=23)

# Обучаем наш классификатор; train_data.target будет содержать числа, присвоенные каждой категории в обучающих данных
clf = knn.fit(X_train_tfidf, target)

# Вводимые данные для предсказания их классов в заданных категориях
docs_new = ['I have a Harley Davidson and Yamaha.', 'I have a GTX 1050 GPU']
# Формируем вектор признаков для нашего ввода
X_new_counts = count_vect.transform(docs_new)
# Мы вызываем transform вместо fit_transform, потому что он уже был обучен
X_new_tfidf = tfidf_transformer.transform(X_new_counts)
```

```python
predicted = clf.predict(X_new_tfidf)
```

```python
from collections import Counter

def for_sampling(docs):

  n_neighbors = 20
  ans = 0
  name = []
  # Формируем вектор признаков для нашего ввода
  X_new_counts = count_vect.transform(docs)
  # Мы вызываем transform вместо fit_transform, потому что он уже был обучен
  X_new_tfidf = tfidf_transformer.transform(X_new_counts)
  while ans == 0:

    distances, indices = clf.kneighbors(X_new_tfidf, n_neighbors=n_neighbors)
    closest_articles_indices = indices.flatten()
    closest_articles_distances = distances.flatten()

    counter = Counter(closest_articles_distances)
    for i in counter.most_common():
      print(i)
      if i[0] == 1:
        continue
      elif i[1] != 1:
        ans = i[0]
        break
      else:
        n_neighbors += 20
        ans = 0
        break
      print(ans)
    if ans != 0:
      pos = np.where(closest_articles_distances==ans)
      name = closest_articles_indices[pos[0]]
      print(pos[0])
      name = data[name[0]]
  return name
```
